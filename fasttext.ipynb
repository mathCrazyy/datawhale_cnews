```python
import fasttext
```

```python
import pandas as pd
from sklearn.metrics import f1_score
```

### 生成fasttext训练可用的文本

```python
#train_df=pd.read_csv("/data/aig/public/wuchunsheng/NLP/competition/data/cnews/train_set.csv",sep="\t",nrows=15000)
train_df=pd.read_csv("/data/aig/public/wuchunsheng/NLP/competition/data/cnews/train_set.csv",sep="\t")
train_df["label_ft"]="__label__"+train_df["label"].astype(str)
train_df[["text","label_ft"]].iloc[:-5000].to_csv("/data/aig/public/wuchunsheng/NLP/competition/data/cnews/train_ft.csv",index=None,header=None,sep="\t")
```

### fasttext训练

```python
model=fasttext.train_supervised("/data/aig/public/wuchunsheng/NLP/competition/data/cnews/train_ft.csv",lr=1.0,wordNgrams=2,verbose=2,minCount=1,epoch=25,loss="hs")
```

```python
val_pred=[model.predict(x)[0][0].split("__")[-1]for x in train_df.iloc[-5000:]["text"]]
print(f1_score(train_df["label"].values[-5000:].astype(str),val_pred,average="macro"))
```

### gridsearch

```python
from sklearn.model_selection import GridSearchCV
param_grid=[
    {
        "lr":[0.01,0.1,1],
        "wordNgrams":[2,3],
        "verbose":[2,4,6],
        "minCount":[1,5,10],
    }
]
```

```python
model=fasttext.train_supervised("/data/aig/public/wuchunsheng/NLP/competition/data/cnews/train_ft.csv")
grid_search=GridSearchCV(model,param_grid,cv=5)
#grid_search.fit()
```

```python
grid_search.fit()
```
